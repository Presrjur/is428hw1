```python
import pandas as pd

# Load the dataset from the CSV file
df = pd.read_csv("Baltimore911.csv")

# Clean the "Inside/Outside" column
df["Inside/Outside"] = df["Inside/Outside"].apply(lambda x: "I" if x == "Inside" else "O" if x == "Outside" else x)

# Save the cleaned dataset back to a new CSV file
df.to_csv("Cleaned_Baltimore911.csv", index=False)

print("Cleaning complete. The cleaned dataset is saved as 'Cleaned_Baltimore911.csv'.")
```

```python
import pandas as pd
import numpy as np

# Load the cleaned dataset
df = pd.read_csv("Cleaned_Baltimore911.csv")

# Create a profile table
profile_table = pd.DataFrame(columns=["Column Name", "Data Type", "Numeric/Discrete", "Unique Values", "Max", "Min", "Mean", "Median", "Std Dev"])

# Iterate through each column in the dataset
for column_name in df.columns:
    column_data = df[column_name]
    data_type = str(column_data.dtypes)

    if np.issubdtype(column_data, np.number):
        # Numeric column
        numeric_discrete = "Numeric"
        unique_values = len(column_data.unique())
        max_value = column_data.max()
        min_value = column_data.min()
        mean_value = column_data.mean()
        median_value = column_data.median()
        std_dev_value = column_data.std()
    else:
        # Discrete column
        numeric_discrete = "Discrete"
        unique_values = len(column_data.unique())
        max_value = ""
        min_value = ""
        mean_value = ""
        median_value = ""
        std_dev_value = ""

    profile_table = profile_table.append({
        "Column Name": column_name,
        "Data Type": data_type,
        "Numeric/Discrete": numeric_discrete,
        "Unique Values": unique_values,
        "Max": max_value,
        "Min": min_value,
        "Mean": mean_value,
        "Median": median_value,
        "Std Dev": std_dev_value
    }, ignore_index=True)

# Calculate the percentage of rows with missing values
missing_percentage = (df.isnull().sum().sum() / (df.shape[0] * df.shape[1])) * 100

# Print the profile table and missing percentage
print(profile_table)
print(f"\nPercentage of rows with missing values: {missing_percentage:.2f}%")
```

```python
import pandas as pd
import numpy as np

# Load the cleaned dataset
df = pd.read_csv("Cleaned_Baltimore911.csv")

# Create empty lists to store data for the profile table
column_names = []
data_types = []
numeric_discrete = []
unique_values = []
max_values = []
min_values = []
mean_values = []
median_values = []
std_dev_values = []
most_common_values = []

# Iterate through each column in the dataset
for column_name in df.columns:
    column_data = df[column_name]
    data_type = str(column_data.dtypes)

    if np.issubdtype(column_data, np.number):
        # Numeric column
        numeric_discrete_val = "Numeric"
        unique_val = len(column_data.unique())
        max_val = column_data.max()
        min_val = column_data.min()
        mean_val = column_data.mean()
        median_val = column_data.median()
        std_dev_val = column_data.std()
        most_common_val = None  # Numeric columns don't have a 'most common' value
    else:
        # Discrete column
        numeric_discrete_val = "Discrete"
        unique_val = len(column_data.unique())
        max_val = None
        min_val = None
        mean_val = None
        median_val = None
        std_dev_val = None
        most_common_val = column_data.mode().values[0]  # Get the most common value

    # Append data to respective lists
    column_names.append(column_name)
    data_types.append(data_type)
    numeric_discrete.append(numeric_discrete_val)
    unique_values.append(unique_val)
    max_values.append(max_val)
    min_values.append(min_val)
    mean_values.append(mean_val)
    median_values.append(median_val)
    std_dev_values.append(std_dev_val)
    most_common_values.append(most_common_val)

# Create a profile table using pandas.concat
profile_table = pd.DataFrame({
    "Column Name": column_names,
    "Data Type": data_types,
    "Numeric/Discrete": numeric_discrete,
    "Unique Values": unique_values,
    "Max": max_values,
    "Min": min_values,
    "Mean": mean_values,
    "Median": median_values,
    "Std Dev": std_dev_values,
    "Most Common Value": most_common_values
})

# Calculate the number of rows and columns in the dataset
num_rows, num_columns = df.shape

# Add rows and columns information to the profile table
profile_table.loc[-1] = ["Number of Rows", "", "", "", "", "", "", "", "", ""]
profile_table.loc[-2] = ["Number of Columns", "", "", "", "", "", "", "", "", ""]
profile_table.iloc[-1, 1] = num_rows
profile_table.iloc[-2, 1] = num_columns

# Calculate the percentage of rows with missing values
missing_percentage = (df.isnull().sum().sum() / (num_rows * num_columns)) * 100

# Print the profile table and missing percentage
print(profile_table)
print(f"\nPercentage of rows with missing values: {missing_percentage:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned dataset
df = pd.read_csv("Cleaned_Baltimore911.csv")

# Group the data by the "Description" column (crime type) and count occurrences
crime_type_counts = df["Description"].value_counts()

# Create a bar plot
plt.figure(figsize=(12, 6))
crime_type_counts.plot(kind="bar")
plt.title("Distribution of Crime Types")
plt.xlabel("Crime Type")
plt.ylabel("Count")
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned dataset
df = pd.read_csv("Cleaned_Baltimore911.csv")

# Convert the "CrimeDate" column to datetime format
df["CrimeDate"] = pd.to_datetime(df["CrimeDate"], format="%m/%d/%Y")

# Extract year, month, and day of the week from the "CrimeDate" column
df["Year"] = df["CrimeDate"].dt.year
df["Month"] = df["CrimeDate"].dt.month
df["DayOfWeek"] = df["CrimeDate"].dt.day_name()

# Create subplots for year, month, and day-of-week distributions
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# Yearly distribution
yearly_counts = df.groupby("Year")["Description"].value_counts().unstack(fill_value=0)
yearly_counts.plot(kind="bar", stacked=True, ax=axes[0])
axes[0].set_title("Yearly Crime Type Distribution")
axes[0].set_ylabel("Count")

# Monthly distribution
monthly_counts = df.groupby("Month")["Description"].value_counts().unstack(fill_value=0)
monthly_counts.plot(kind="bar", stacked=True, ax=axes[1])
axes[1].set_title("Monthly Crime Type Distribution")
axes[1].set_ylabel("Count")

# Day of the week distribution
day_of_week_counts = df.groupby("DayOfWeek")["Description"].value_counts().unstack(fill_value=0)
day_of_week_counts = day_of_week_counts.reindex(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
day_of_week_counts.plot(kind="bar", stacked=True, ax=axes[2])
axes[2].set_title("Day of the Week Crime Type Distribution")
axes[2].set_ylabel("Count")

# Adjust subplot spacing
plt.tight_layout()

# Show the plots
plt.show()
```
